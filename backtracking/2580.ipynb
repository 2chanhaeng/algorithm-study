```python
def solution():
    import sys

    sudoku = []
    blanks = []
    sudoku_T = {col: set() for col in range(9)}
    squares = {(row, col): set() for row in range(3) for col in range(3)}
    for x, input in enumerate(sys.stdin.read().rstrip().split("\n")):
        line = list(map(int, input[::2]))
        sudoku.append(line)
        for y, cell in enumerate(line):
            sudoku_T[y].add(cell)
            squares[(x // 3, y // 3)].add(cell)
            if cell == 0:
                blanks.append((x, y))

    onetonine = set(range(1, 10))

    def solve_sudoku(depth):
        if depth == 0:
            return sudoku
        row, col = blanks[depth - 1]
        rcp3 = (row // 3, col // 3)
        able = onetonine - (set(sudoku[row]) | sudoku_T[col] | squares[rcp3])
        for i in able:
            sudoku[row][col] = i
            sudoku_T[col].add(i)
            squares[rcp3].add(i)
            result = solve_sudoku(depth - 1)
            if result:
                return result
            sudoku_T[col].discard(i)
            squares[rcp3].discard(i)
        else:
            sudoku[row][col] = 0
            return False

    for line in solve_sudoku(len(blanks)):
        print(*line)


solution()
```

```python
def solution():
    import sys

    sudoku = []
    blank = []
    for row, input in enumerate(sys.stdin):
        sudoku.append(line := input.split())
        blank.extend([(row, col) for col, cell in enumerate(line) if cell == "0"])
    blanks = len(blank)
    nums = "123456789"

    def is_able(x, row, col):
        if x in sudoku[row]:
            return False
        for line in sudoku:
            if x == line[col]:
                return False
        rp3 = row // 3 * 3
        cp3 = col // 3 * 3
        for r in range(rp3, rp3 + 3):
            for c in range(cp3, cp3 + 3):
                if x == sudoku[r][c]:
                    return False
        return True
    
    def backtracking(n = 0):
        if n == blanks:
            return True
        row, col = blank[n]
        for i in nums:
            if is_able(i, row, col):
                sudoku[row][col] = i
                if backtracking(n + 1):
                    return True
                sudoku[row][col] = 0
        return False

    backtracking()
    for i in sudoku: print(*i)


solution()
```
