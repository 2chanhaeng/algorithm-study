## 풀이과정

### 첫번째 시도

풀이과정

```python
from collections import Counter as C
import sys

rows, cols, default, *heights = map(int, open(0).read().split())
total = sum(heights)
blocks_by_heights = C(heights)
heights_set = sorted(blocks_by_heights.keys())
M, m = heights_set[-1], heights_set[0]
min_time = total * 2
max_floor = m
for floor in range(M, m - 1, -1):
    time = 0
    inven = default
    for height in heights_set[::-1]:
        if height < floor:
            blocks = blocks_by_heights[height] * (floor - height)
            time += blocks
            inven -= blocks

        elif height > floor:
            blocks = blocks_by_heights[height] * (height - floor)
            time += 2 * blocks
            inven += blocks
    if inven >= 0 and min_time > time:
        min_time = time
        max_floor = floor
print(min_time, max_floor)
```

## 해답

```python
from collections import Counter as C
```

```python
def solution(open):
    rows, cols, default, *heights = map(int, open(0).read().split())
    total = sum(heights)
    blocks_by_heights = C(heights)
    heights_set = sorted(blocks_by_heights.keys())
    print(heights_set)
    M, m = heights_set[-1], heights_set[0]
    min_time = total * 2 + 1
    max_floor = m
    for floor in range(M, m - 1, -1):
        print(f"{floor = }")
        time = 0
        inven = default
        for height in heights_set[::-1]:
            print(f" {height = }")
            if height < floor:
                blocks = blocks_by_heights[height] * (floor - height)
                time += blocks
                inven -= blocks
                print(f"  {blocks = } {time = }")

            elif height > floor:
                blocks = blocks_by_heights[height] * (height - floor)
                time += 2 * blocks
                inven += blocks
                print(f"  {blocks = } {time = }")
        print(f"{min_time = } {time = } {inven = }")
        if inven >= 0 and min_time > time:
            min_time = time
            max_floor = floor
            print(f" {min_time = } {max_floor = }")
    print(min_time, max_floor)
```

## 예제

```python
# 백준 문제 풀이용 예제 실행 코드
from inspect import signature


def solution_example(input_=None, solution=solution):
    params = signature(solution).parameters
    if input_ == None:
        solution()
    elif "open" in params:

        class open:
            def __init__(self, x):
                self.read = lambda: input_

            def __iter__(self):
                return iter(input_.split("\n"))

        solution(open)
    elif "input" in params:
        input_ = iter(input_.split("\n"))
        input = input_.__next__
        solution(input)
    else:
        raise NameError(
            f"solution 함수가 input, open 등의 매개변수를 받지 않습니다.\nsolution 의 매개변수: {params}"
        )


# solution_example("""""")
# solution_example(read("fn").read())
```
