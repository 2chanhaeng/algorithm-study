## 풀이과정


### 첫번째 시도


#### 풀이과정


```python
def diffs(ns: list[int]) -> list[int]:
    return [m - n - 1 for n, m in zip(ns, ns[1:])]


def solution():
    import sys

    inputs = (map(int, line.split()) for line in sys.stdin.read().strip().split("\n"))
    w, h, n = next(inputs)
    if n == 0:
        print(w * h)
        return
    ws, hs = (sorted(i) for i in zip(*inputs))
    max_w = max([ws[0] - 1, w - ws[-1], *diffs(ws)])
    max_h = max([hs[0] - 1, h - hs[-1], *diffs(hs)])
    print(max_w * max_h)


solution()
```
