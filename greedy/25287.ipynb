## 풀이과정


### 첫번째 시도


#### 풀이과정


```python
def solve(nums: list[int]) -> str:
    l = len(nums)
    c = 0
    for n in nums:
        m = l - n + 1
        if m < n:
            m, n = n, m
        if m < c:
            return "NO"
        c = n if c <= n else m
    return "YES"


def solution():
    import sys

    inputs = sys.stdin.read().strip().split("\n")[2::2]
    for line in inputs:
        print(solve(list(map(int, line.split()))))


solution()
```
