## [AFC 윔블던](https://www.acmicpc.net/problem/4299)

```python
a, b = map(int, input().split())
w, l = (a + b) // 2, (a - b) // 2
if a >= b and ((a + b) % 2) + ((a - b) % 2)== 0 and w >= l >= 0:
    print(w, l)
else:
    print(-1)
```

## [Dedupe](https://www.acmicpc.net/problem/5357)

```python
def depude(string):
    yield (p := string[0])
    for s in string[1:]:
        if s != p:
            yield (p := s)


print("\n".join("".join(depude(input())) for _ in range(int(input()))))
```

## [Football Team](https://www.acmicpc.net/problem/5358)

```python
def replace(string):
    for c in string:
        match c:
            case "i":
                yield "e"
            case "e":
                yield "i"
            case "I":
                yield "E"
            case "E":
                yield "I"
            case _:
                yield c
for string in open(0):
    print("".join(replace(string)))
```

## [타임 카드](https://www.acmicpc.net/problem/5575)

```python
from datetime import datetime as dt
for _ in range(3):
    h, m, s, H, M, S = map(int, input().split())
    secs = int((dt(1, 1, 1, H, M, S) - dt(1, 1, 1, h, m, s)).total_seconds())
    print(secs // 3600, secs % 3600 // 60, secs % 60)
```
