## 풀이과정

### 첫번째 시도

풀이과정

```python
def solution(open=open):
    input = iter(open(0).read().split("\n")).__next__
    N = int(input())
    graph = [list(map(int, input()[::2])) for _ in range(N)]
    for mid in range(N):
        for start in range(N):
            for end in range(N):
                if graph[start][mid] & graph[mid][end]:
                    graph[start][end] = 1
    for node in graph:
        print(" ".join(map(str, node)))


solution()
```

## 해답

```python
def search(row, visited, N, graph, depth=0):
    print(" " * depth + "A", row)
    for i, is_conned in enumerate(bin(graph[row])[2:].zfill(N)):
        if is_conned == "0":
            continue
        print(" " * depth + " B", row, i)
        if i not in visited:
            visited.add(i)
            search(i, visited, N, graph, depth + 1)
            graph[row] |= graph[i]
        else:
            graph[row] = graph[i] = graph[i] | graph[row] | (1 << i)
    print(" " * depth + "C", graph[row])
```

```python
def solution(open=open):
    input = iter(open(0).read().split("\n")).__next__
    N = int(input())
    graph = [int(input()[::2], 2) for _ in range(N)]
    for i in range(N):
        visited = {i}
        search(i, visited, N, graph)
        print(f"\n{i = }, {visited = }")
        for node in graph:
            print(" ".join(bin(node)[2:].zfill(N)))
        print()
```
