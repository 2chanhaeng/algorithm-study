### 세번째 시도


```python
# Reference: https://www.acmicpc.net/source/19521628
def solution():
    import sys
    from itertools import accumulate, count
    from operator import mul

    DIMS = [i for i in map(int, sys.stdin.readline().split()) if i > 1]
    NEARS = list(accumulate(DIMS[:-1], mul, initial=1))
    tomatoes = bytearray(map(lambda i: int(i) + 1, sys.stdin.read().split()))

    def coord(n):
        for i in DIMS:
            n, r = divmod(n, i)
            yield r

    def init(tomatoes):
        for i, t in enumerate(tomatoes):
            if t > 1:
                tomatoes[i] = 0
                yield i

    def neighbor(riped):
        for index in riped:
            for c, d, n in zip(coord(index), DIMS, NEARS):
                if c + 1 < d:
                    near = index + n
                    if tomatoes[near]:
                        tomatoes[near] = 0
                        yield near
                if 0 < c:
                    near = index - n
                    if tomatoes[near]:
                        tomatoes[near] = 0
                        yield near

    def solve(riped):
        for counter in count():
            riped = list(neighbor(riped))
            if not riped:
                return -1 if any(tomatoes) else counter

    print(solve(init(tomatoes)))


solution()
```

```python
def z(k, x):
    m = 1
    s = 0
    for i, j in zip(k[::-1], x[::-1]):
        s += i * m
        m *= j
    return s


print(z([0, 0, 0], [3, 2, 1]))
print(z([0, 1, 0], [3, 2, 1]))
print(z([1, 0, 0], [3, 2, 1]))
print(z([1, 1, 0], [3, 2, 1]))
print(z([2, 0, 0], [3, 2, 1]))
print(z([2, 1, 0], [3, 2, 1]))
```
