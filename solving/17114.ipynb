```python
def solution(input):
    m, n, o, p, q, r, s, t, u, v, w = list(map(int, input().split()))
    # m ~ w 까지 입력받습니다.
    tomatoes = {}
    # 격자간 토마토의 상태를 저장할 딕셔너리를 생성합니다.
    for W in range(w):
        for V in range(v):
            for U in range(u):
                for T in range(t):
                    for S in range(s):
                        for R in range(r):
                            for Q in range(q):
                                for P in range(p):
                                    for O in range(o):
                                        for N in range(n):
                                            for M, i in zip(range(m), map(int, input().split())):
                                                tomatoes[(M, N, O, P, Q, R, S, T, U, V, W)] = i
                                                # 마지막에만 입력받은 숫자를 상태로써 저장합니다.
    # 키는 토마토의 좌표, 값은 토마토의 상태입니다.
    mm, nm, om, pm, qm, rm, sm, tm, um, vm, wm = m-1, n-1, o-1, p-1, q-1, r-1, s-1, t-1, u-1, v-1, w-1
    # 매번 1을 빼는 것도 시간이 걸릴까봐 변수로 지정했습니다.
    def side(points, empty):
        # 주변 토마토를 찾는 함수입니다.
        sides = set()
        # 최종적으로 반환할 집합입니다.
        for point in points:
            M, N, O, P, Q, R, S, T, U, V, W = point
            # 0 보다 크면, +1 좌표를 추가합니다.
            # 최대보다 작으면, -1 좌표를 추가합니다.
            if M < mm: sides.add((M + 1, N, O, P, Q, R, S, T, U, V, W))
            if M > 0:  sides.add((M - 1, N, O, P, Q, R, S, T, U, V, W))
            if N < nm: sides.add((M, N + 1, O, P, Q, R, S, T, U, V, W))
            if N > 0:  sides.add((M, N - 1, O, P, Q, R, S, T, U, V, W))
            if O < om: sides.add((M, N, O + 1, P, Q, R, S, T, U, V, W))
            if O > 0:  sides.add((M, N, O - 1, P, Q, R, S, T, U, V, W))
            if P < pm: sides.add((M, N, O, P + 1, Q, R, S, T, U, V, W))
            if P > 0:  sides.add((M, N, O, P - 1, Q, R, S, T, U, V, W))
            if Q < qm: sides.add((M, N, O, P, Q + 1, R, S, T, U, V, W))
            if Q > 0:  sides.add((M, N, O, P, Q - 1, R, S, T, U, V, W))
            if R < rm: sides.add((M, N, O, P, Q, R + 1, S, T, U, V, W))
            if R > 0:  sides.add((M, N, O, P, Q, R - 1, S, T, U, V, W))
            if S < sm: sides.add((M, N, O, P, Q, R, S + 1, T, U, V, W))
            if S > 0:  sides.add((M, N, O, P, Q, R, S - 1, T, U, V, W))
            if T < tm: sides.add((M, N, O, P, Q, R, S, T + 1, U, V, W))
            if T > 0:  sides.add((M, N, O, P, Q, R, S, T - 1, U, V, W))
            if U < um: sides.add((M, N, O, P, Q, R, S, T, U + 1, V, W))
            if U > 0:  sides.add((M, N, O, P, Q, R, S, T, U - 1, V, W))
            if V < vm: sides.add((M, N, O, P, Q, R, S, T, U, V + 1, W))
            if V > 0:  sides.add((M, N, O, P, Q, R, S, T, U, V - 1, W))
            if W < wm: sides.add((M, N, O, P, Q, R, S, T, U, V, W + 1))
            if W > 0:  sides.add((M, N, O, P, Q, R, S, T, U, V, W - 1))
        return sides & empty
        # 그 중에서 익지 않은 토마토의 좌표만 가져옵니다.
    empty, changed = set(), set()
    # empty 는 익지 않은 토마토의 좌표가 저장될 집합입니다.
    # changed 는 비어있거나 바뀐 토마토의 좌표가 저장될 집합입니다.
    tochange = set()
    # tochange 는 상태가 바뀔 토마토의 좌표가 저장될 집합입니다.
    # 다만 0번째에 들어가기 전, 즉 -1번째 날(가정)에는 익은 토마토들의 자리를 저장합니다.
    # 그래야 0번째 날에 익은 토마토가 되기 때문입니다.
    for key, value in tomatoes.items():
        # 토마토의 좌표와 상태를 가져옵니다.
        if value == 1:
            # 익은 토마토라면,
            changed.add(key)
            # 좌표를 익은 토마토 좌표와
            tochange.add(key)
            # 바뀔 토마토 좌표에 추가합니다.
        elif value:
            # 비어있다면,
            changed.add(key)
            # 좌표를 빈 자리 좌표에 추가합니다.
        else:
            # 둘 다 아니라면,
            empty.add(key)
            # 좌표를 익지 않은 토마토 좌표에 추가합니다.
    for days in range(m*n*o*p*q*r*s*t*u*v*w):
        # 날을 세는 반복문입니다.
        # 최댓값을 격자의 칸 수로 두었습니다.
        tochange = side(tochange, empty)
        # 오늘 익을 토마토의 집합을 전날 바뀔 토마토의 집합으로 구합니다.
        empty -= tochange
        # 바뀔 토마토의 집합을 익지 않은 토마토 집합에서 제거합니다.
        changed |= tochange
        # 바뀐 토마토의 집합에 익은 토마토 좌표에 추가합니다.
        if not tochange: break
        # 바뀔 토마토의 집합이 비었다면 반복문을 종료합니다.
    print(-1 if empty else days)
    # 익지 않은 토마토가 남았다면 -1을 출력하고, 아니라면 며칠이 걸렸는지 출력합니다.
import sys
solution(lambda: sys.stdin.readline().rstrip())
```

```python
def solution2(input):
    x = [w ,v ,u ,t ,s ,r ,q ,p ,o ,n ,m] = list(map(int, input().split()))
    pivots = list(product(*(list(range(i)) for i in x[::-1])))
    tomatoes = {}
    m, N = x[0], prod(x[1:])
    for k in range(N):
        for i, j in zip(pivots[k * m:(k + 1) * m], list(map(int, input().split()))):
            tomatoes[i] = j
    def side(points, empty):
        sides = set()
        for point in points:
            M, N, O, P, Q, R, S, T, U, V, W = point
            if M < m - 1:   sides.add((M + 1, N, O, P, Q, R, S, T, U, V, W))
            if M > 0:       sides.add((M - 1, N, O, P, Q, R, S, T, U, V, W))
            if N < n - 1:   sides.add((M, N + 1, O, P, Q, R, S, T, U, V, W))
            if N > 0:       sides.add((M, N - 1, O, P, Q, R, S, T, U, V, W))
            if O < o - 1:   sides.add((M, N, O + 1, P, Q, R, S, T, U, V, W))
            if O > 0:       sides.add((M, N, O - 1, P, Q, R, S, T, U, V, W))
            if P < p - 1:   sides.add((M, N, O, P + 1, Q, R, S, T, U, V, W))
            if P > 0:       sides.add((M, N, O, P - 1, Q, R, S, T, U, V, W))
            if Q < q - 1:   sides.add((M, N, O, P, Q + 1, R, S, T, U, V, W))
            if Q > 0:       sides.add((M, N, O, P, Q - 1, R, S, T, U, V, W))
            if R < r - 1:   sides.add((M, N, O, P, Q, R + 1, S, T, U, V, W))
            if R > 0:       sides.add((M, N, O, P, Q, R - 1, S, T, U, V, W))
            if S < s - 1:   sides.add((M, N, O, P, Q, R, S + 1, T, U, V, W))
            if S > 0:       sides.add((M, N, O, P, Q, R, S - 1, T, U, V, W))
            if T < t - 1:   sides.add((M, N, O, P, Q, R, S, T + 1, U, V, W))
            if T > 0:       sides.add((M, N, O, P, Q, R, S, T - 1, U, V, W))
            if U < u - 1:   sides.add((M, N, O, P, Q, R, S, T, U + 1, V, W))
            if U > 0:       sides.add((M, N, O, P, Q, R, S, T, U - 1, V, W))
            if V < v - 1:   sides.add((M, N, O, P, Q, R, S, T, U, V + 1, W))
            if V > 0:       sides.add((M, N, O, P, Q, R, S, T, U, V - 1, W))
            if W < w - 1:   sides.add((M, N, O, P, Q, R, S, T, U, V, W + 1))
            if W > 0:       sides.add((M, N, O, P, Q, R, S, T, U, V, W - 1))
        return sides & empty
    days = 0
    empty, changed = set(), set()
    getside = set()
    for key, value in tomatoes.items():
        [empty, changed][value].add(key)
        if value == 1:
            getside.add(key)
    tochange = side(getside, empty)
    getside.clear()
    for _ in range(m*n*o*p*q*r*s*t*u*v*w):
        if not tochange:
            break
        empty -= tochange
        changed |= tochange
        tochange = side(tochange, empty)
        days += 1
    if empty:
        print(-1)
    else:
        print(days)
```
